TOKENIZER FOR LLM (Large Language Models) : SIMPLE TOKENIZER IMPLEMENTATION


CONVERTING TEXT INTO TOKENS 

```python
import re #importing the regular expression module to use for splitting the text into tokens. Powerful then just spliting on whitespace. It allows us to split the text based on punctuation and other delimiters as well.
```

```python
text = "The world is the stage, and all the men and women merely players."  #input text
y= re.split(r'([,.:;?_!"()\']|--|\s)', text) #split the text into tokens based on punctuation and whitespace
y=[item.strip() for item in y if item.strip()] #remove any empty tokens and strip whitespace from the tokens

print(y)
print(len(y)) #length of the list of tokens
```

CONVERTING TOKENS INTO TOKENS-ID


VOCABALARY GENERATION FOR MODEL - SET OF UNIQUE TOKENS (VOCABALARY)

```python
all_words = sorted(set(y))
vocab_size = len(all_words)

print(vocab_size)
print(all_words)
```

```python
vocab = {token:integer for integer,token in enumerate(all_words)}
print(vocab)
```

```python
class SimpleTokenizerV1:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s,i in vocab.items()}
    
    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
                                
        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]
        ids = [self.str_to_int[s] for s in preprocessed]


        return ids
        
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text
```

```python
tokenizer = SimpleTokenizerV1(vocab)

text = "The world is the stage, "
ids = tokenizer.encode(text)
text_decoded = tokenizer.decode(ids)
print(text_decoded)
print(ids)
```
